```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# --- CONFIGURATION ---
# Replace these filenames with the actual names of the CSVs you downloaded
FILE_SLEEP = "../data/sleep_health_lifestyle_dataset.csv"
FILE_MENTAL = "../data/student_stress_sleep_screen.csv"
FILE_PROD = "../data/productivity_dataset.csv"
```

```python
# --- 1. LOAD DATA ---
try:
    df_sleep = pd.read_csv(FILE_SLEEP)
    df_mental = pd.read_csv(FILE_MENTAL)
    df_prod = pd.read_csv(FILE_PROD)
    print("✅ All Datasets Loaded Successfully!")
except FileNotFoundError as e:
    print(f"❌ Error: {e}")
    print("Make sure you put the CSV files in the 'app/ml_engine/data' folder!")
```

```python
# --- 2. INSPECT THE BODY (Sleep Data) ---
print("\n--- SLEEP DATASET PREVIEW ---")
display(df_sleep.head())
print(f"Shape: {df_sleep.shape}")
```

```python
# Visualization 1: Does Sleep Duration affect Stress?
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_sleep, x='Sleep Duration (hours)', y='Stress Level (scale: 1-10)', hue='BMI Category')
plt.title("Correlation: Sleep vs. Stress")
plt.show()
```
